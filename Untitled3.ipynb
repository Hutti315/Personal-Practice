```python
import pandas as pd

data = {
    'order_id': ['HD001', 'HD002', 'HD003', 'HD004', 'HD005', 'HD006', 'HD007', 'HD008', 'HD009', 'HD010'],
    'date': ['2026-01-05', '2026-01-12', '2026-02-02', '2026-02-15', '2026-03-01', '2026-03-22', '2026-04-10', '2026-04-29', '2026-05-05', '2026-05-19'],
    'product': ['iPhone 15', 'Logitech Mouse', 'iPhone 15', 'MacBook Air', 'Logitech Mouse', 'MacBook Air', 'iPhone 15', 'AirPods Pro', 'AirPods Pro', 'MacBook Air'],
    'category': ['Điện thoại', 'Phụ kiện', 'Điện thoại', 'Máy tính', 'Phụ kiện', 'Máy tính', 'Điện thoại', 'Phụ kiện', 'Phụ kiện', 'Máy tính'],
    'quantity': [1, 5, 2, 1, 10, 2, 1, 3, 2, 1],
    'price_each': ['23,000,000', '850,000', '22,500,000', '26,000,000', '800,000', '25,500,000', '23,000,000', '5,000,000', '4,900,000', '26,000,000']
}

df_sales = pd.DataFrame(data)
df_sales.to_csv('sales_data.csv', index=False)
print("Đã tạo file sales_data.csv thành công!")
```

```python
df_sales = pd.read_csv("sales_data.csv")
df_sales.info()
```

```python
df_sales["price_each"] = df_sales["price_each"].str.replace(",", "")
df_sales["price_each"] = pd.to_numeric(df_sales["price_each"])
df_sales["total_revenue"] = df_sales["quantity"] * df_sales["price_each"]
```

```python
df_sales.head()
```

```python
revenue_by_product = df_sales.groupby("product")["total_revenue"].sum()
revenue_by_product = revenue_by_product.sort_values(ascending=False)

print(revenue_by_product)
```

```python
import matplotlib.pyplot as plt
revenue_by_product.plot(kind="bar", color="#50C878", edgecolor="black")

plt.title("Doanh Thu Theo Từng Sản Phẩm (5 Tháng Đầu Năm)", fontsize=14, fontweight='bold')
plt.xlabel("Sản Phẩm")
plt.ylabel("Doanh Thu (VNĐ)")
plt.xticks(rotation=45, color="red")
plt.tight_layout()
plt.show()
```

```python
df_revenue = revenue_by_product.reset_index()
df_revenue.columns = ["Sản phẩm", "Tổng doanh thu (VNĐ)"]

with pd.ExcelWriter("Bao_cao_Doanh_thu.xlsx", engine="openpyxl") as writer:
    df_revenue.to_excel(writer, index=False, sheet_name="Doanh thu")

    workbook = writer.book
    worksheet = writer.sheets["Doanh thu"]
    for row in range(2, len(df_revenue) +2):
        worksheet[f"B{row}"].number_format = '#,##0'
    print("Đã xuất file Bao_cao_Doanh_thu.xlsx !")
```
