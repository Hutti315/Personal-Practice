```python
import matplotlib.pyplot as plt
top10 = df.sort_values(by="points", ascending=False).head(10)
plt.bar(top10["team"], top10["points"], color="skyblue")

top10.to_excel("Top_10_FIFA.xlsx", index=False)

plt.title("Top 10 Đội Bóng Có Điểm Số Cao Nhất Thế Giới")
plt.xlabel("Tên Đội Bóng", fontsize=12)
plt.ylabel("Điểm Số", fontsize=12)
plt.xticks(rotation=45, color="red")

plt.tight_layout()

plt.savefig("Bieu_do_Top10.png", dpi=300)

plt.show()
```
