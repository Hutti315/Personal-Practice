```python
import pandas as pd

food_cost_pdpo = 3000
sheep_quantity = 100
worker_quantity = 1
medicine_cost_opm = 50000
worker_salary = 3500000
electricity_water_bill = 2000000
```

```python
total_foodcostmonthly = food_cost_pdpo * sheep_quantity * 30
VM = medicine_cost_opm * sheep_quantity
WC = worker_salary * worker_quantity
```

```python
Total_cost = total_foodcostmonthly + VM + WC
print(f"Total cost to coporate sheep farming in once month: {Total_cost:,.0f} VNĐ")
```

```python
wool_ratio = 0.8
wool_weight_per_sheep = 4
wool_price = 120000
meat_sheep_sold = 30
meat_price = 5000000
```

```python
revenue_wool = wool_ratio * wool_weight_per_sheep * wool_price * sheep_quantity
revenue_meat = meat_sheep_sold * meat_price
total_revenue = revenue_wool + revenue_meat

print(f"Total revenue from sheep farming in once month is: {total_revenue:,.0f} VNĐ")
```

```python
total_cost_year = Total_cost * 12
net_profit = total_revenue - total_cost_year
net_profit_margin = (net_profit / total_revenue) * 100

print(f"total cost in 1 year: {total_cost_year:,.0f} VNĐ")
print(f"net profit: {net_profit:,.0f} VNĐ")
print(f"net profit margin: {net_profit_margin:.2f} VNĐ")

if net_profit_margin > 20:
    print("Conclusion: Farming is good at operating, highly net profit margin")
else:
    print("Conclusion: Warning!! Farming is lossed. IT's must saving cost open right now")
```
