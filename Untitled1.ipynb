```python
import pandas as pd
```

```python
df_spend = pd.read_excel("Chi_tieu_hang_ngay.xlsx", header=1)
df_spend = df_spend.loc[:, ~df_spend.columns.str.contains('^Unnamed')]
df_spend["Số tiền"] = df_spend["Số tiền"].astype(str).str.replace('.', '', regex=False)
df_spend["Số tiền"] = pd.to_numeric(df_spend["Số tiền"], errors='coerce')
df_spend["Ngày"] = df_spend["Ngày"].fillna(0)
pd.options.display.float_format = '{:, .0f}'.format`
df_spend
```

```python
sumary_spend = df_spend.groupby("Danh mục")["Số tiền"].sum().reset_index()
sumary_spend = sumary_spend.sort_values(by="Số tiền", ascending=False)
sumary_spend
```
